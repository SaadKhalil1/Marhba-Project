# **Data visualization**

Now lets take a look at how the sales amount is distributed



```python
plt.figure(figsize=(9, 8))
sns.distplot(df['sales_amount'], color='g', bins=100, hist_kws={'alpha': 0.4});
```

distribution of all of the features by ploting them

```python
df_num = df.select_dtypes(include = ['float64', 'int64'])
df_num.head()
```

```python
df_num.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8); # ; avoid having the matplotlib verbose informations
```

**which features are strongly correlated with Sales Amount**

```python
df_num_corr = df_num.corr()['sales_amount'][:-1] # -1 because the latest row is SalePrice
golden_features_list = df_num_corr[abs(df_num_corr) > 0.5].sort_values(ascending=False)
print("There is {} strongly correlated values with sales amount:\n{}".format(len(golden_features_list), golden_features_list))
```

```python
for i in range(0, len(df_num.columns), 5):
    sns.pairplot(data=df_num,
                x_vars=df_num.columns[i:i+5],
                y_vars=['sales_amount'])
```
